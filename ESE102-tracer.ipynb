### Import packages required for this homework:

```python
import numpy as np                      # Python's numerical and array calculations package https://numpy.org/
import matplotlib.pyplot as plt        # Python's default plotting package https://matplotlib.org/
```

### Specify plotting equations: 

```python
# number of points along tracer edge
n = 128

# angle describing the initial locations of the points
θ = 2*np.pi*np.arange(n+1)/n

# initial locations of points along the tracer edge
x = np.cos(θ)
y = np.sin(θ)
```

### Plot tacer edge: 

```python
fig, ax = plt.subplots(1, 1)
ax.set_aspect(1)             # set plot aspect ratio 
ax.plot(x, y)                   # plot initial x and y from above
ax.set_xlabel("$x$")        # set x axis label
ax.set_ylabel("$y$")
fig.savefig("tracer.pdf")   # save figure into a pdf in the folder you are currently working in
plt.show()                      # show plot
```
